```python
import zipfile
import os

# Unzip the downloaded dataset
with zipfile.ZipFile("default-of-credit-card-clients-dataset.zip", "r") as zip_ref:
    zip_ref.extractall("data")

# List the files in the directory to find the extracted file name
extracted_files = os.listdir("data")
csv_file = [file for file in extracted_files if file.endswith('.csv')][0]

# Read the CSV file into a DataFrame
df = pd.read_csv(os.path.join("data", csv_file))

```

### Modeling

```python
X = df.to_numpy()
X.shape
```

```python
from sklearn.cluster import KMeans

#sum of squared distances
ssd = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k, n_init="auto")
    kmeans.fit(X)
    ssd.append(kmeans.inertia_)

fig = plt.figure(figsize=(10, 3))
plt.plot(range(2, 20), ssd)
plt.xticks(range(2, 20))
plt.ylabel("within distance")
plt.xlabel("k")
plt.grid(True)
plt.title('Elbow curve')
```

```python
from sklearn.cluster import KMeans

# Define number of
k = 3

# Define model
model = KMeans(n_clusters=3)

# Fit model
model.fit(X)
yhat = model.labels_

yhat
```
