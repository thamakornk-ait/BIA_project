```python
import zipfile
import os

# Unzip the downloaded dataset
with zipfile.ZipFile("default-of-credit-card-clients-dataset.zip", "r") as zip_ref:
    zip_ref.extractall("data")

# List the files in the directory to find the extracted file name
extracted_files = os.listdir("data")
csv_file = [file for file in extracted_files if file.endswith('.csv')][0]

# Read the CSV file into a DataFrame
df = pd.read_csv(os.path.join("data", csv_file))

```
